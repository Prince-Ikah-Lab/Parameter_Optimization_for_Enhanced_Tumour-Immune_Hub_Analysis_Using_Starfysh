## Load Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Training Loss Analysis

#### WIthout H&E Information

```python
# Training loss data without H&E information
training_loss_noHE_results = {
    "n_anchors": [10, 20, 60, 120],
    "Window_Size_1": [1623.1469, 1623.1469, 1623.1469, 1623.1469],
    "Window_Size_2": [1624.0984, 1624.0984, 1624.0984, 1624.0984],
    "Window_Size_3": [1621.9249, 1621.9249, 1621.9249, 1621.9249],
    "Window_Size_5": [1620.8449, 1620.8449, 1620.8449, 1620.8449],
    "Window_Size_7": [1621.2543, 1621.2543, 1621.2543, 1621.2543],
    "Window_Size_9": [1621.5243, 1621.5243, 1621.5243, 1621.5243],
    "Window_Size_11": [1621.5282, 1621.5282, 1621.5282, 1621.5282],
}

# Convert the dictionary to a DataFrame
training_loss_noHE = pd.DataFrame(training_loss_noHE_results)
```

```python
training_loss_noHE
```

```python
# Create a heatmap to display the training loss data
plt.figure(figsize=(8, 6))
sns.heatmap(training_loss_noHE.set_index("n_anchors"), annot=True, cmap="Purples", fmt=".4f", cbar=True)
plt.title("Heatmap of Training Loss without H&E information", fontsize=14, pad=25)
plt.ylabel("n_anchors")
plt.tight_layout()

# Display the heatmap
plt.show()
```

```python
# Create DataFrame for the training loss including more window sizes
training_loss_withoutHE_results = {
    "window_size": [1, 2, 3, 5, 7, 9, 11],
    "training_loss": [1623.1469, 1624.0984, 1621.9249, 1620.8449, 1621.2543, 1621.5243, 1621.5282],
}

# Convert the dictionary to a DataFrame
training_loss_withoutHE = pd.DataFrame(training_loss_withoutHE_results)
```

```python
training_loss_withoutHE
```

```python
# Create the line chart with window_size as the x-axis and modified settings
plt.figure(figsize=(10, 6))

# Plot the training loss data
plt.plot(
    training_loss_withoutHE["window_size"],
    training_loss_withoutHE["training_loss"],
    marker='o',
    color="purple"
)

# Add data point values on each point
for x, y in zip(training_loss_withoutHE["window_size"], training_loss_withoutHE["training_loss"]):
    plt.text(x, y, f"{y:.4f}", fontsize=10, ha='center', va='bottom')

# Add labels and title with padding
plt.title("Training Loss Across Window Sizes without H&E information", fontsize=14, pad=25)
plt.xlabel("Window Size", fontsize=12, labelpad=15)
plt.ylabel("Training Loss", fontsize=12, labelpad=15)
plt.xticks(range(1, 12 + 1))

plt.grid(False)
plt.tight_layout()

# Display the plot
plt.show()
```

## Training time analysis without H&E information

```python
# First run Starfysh model training time

# Without H&E information
first_train_noHE_results = {
    "n_anchors": [10, 20, 60, 120],
    "Window_Size_1": [29.00, 24.58, 27.40, 22.37],
    "Window_Size_2": [29.37, 31.15, 21.33, 21.68],
    "Window_Size_3": [36.00, 32.85, 25.43, 23.90],
    "Window_Size_5": [24.77, 19.85, 22.75, 30.35],
}

first_train_noHE = pd.DataFrame(first_train_noHE_results)
```

```python
first_train_noHE
```

```python
# Second run Starfysh model training time

# Without H&E information
second_train_noHE_results = {
    "n_anchors": [10, 20, 60, 120],
    "Window_Size_1": [22.85, 20.65, 21.95, 22.37],
    "Window_Size_2": [22.58, 19.53, 27.62, 21.27],
    "Window_Size_3": [21.62, 19.25, 25.42, 21.20],
    "Window_Size_5": [24.53, 21.50, 25.53, 18.70],
}

second_train_noHE = pd.DataFrame(second_train_noHE_results)
```

```python
second_train_noHE
```

```python
# Third run Starfysh model training time

# Without H&E information
third_train_noHE_results = {
    "n_anchors": [10, 20, 60, 120],
    "Window_Size_1": [20.38, 28.07, 20.50, 20.92],
    "Window_Size_2": [29.37, 20.98, 26.63, 21.53],
    "Window_Size_3": [27.55, 22.22, 28.42, 19.37],
    "Window_Size_5": [21.43, 18.83, 25.48, 25.92],
}

third_train_noHE = pd.DataFrame(third_train_noHE_results)
```

```python
third_train_noHE
```

```python
# Set up a figure for the no H&E heatmaps
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
#fig.suptitle("", fontsize=16)

sns.heatmap(first_train_noHE.set_index('n_anchors'), annot=True, cmap="Blues", fmt=".2f", ax=axs[0])
axs[0].set_title("First Train No H&E", pad=10)

sns.heatmap(second_train_noHE.set_index('n_anchors'), annot=True, cmap="Blues", fmt=".2f", ax=axs[1])
axs[1].set_title("Second Train No H&E", pad=10)

sns.heatmap(third_train_noHE.set_index('n_anchors'), annot=True, cmap="Blues", fmt=".2f", ax=axs[2])
axs[2].set_title("Third Train No H&E", pad=10)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Calculate the mean values of training times across the three tables without H&E image for each cell
mean_table = pd.DataFrame()

# Iterate through all columns except 'n_anchors'
for column in first_train_noHE.columns[1:]:
    mean_table[column] = (first_train_noHE[column] +
                          second_train_noHE[column] +
                          third_train_noHE[column]) / 3

# Add 'n_anchors' column back
mean_table["n_anchors"] = first_train_noHE["n_anchors"]
mean_table_noHE = mean_table[["n_anchors"] + [col for col in mean_table.columns if col != "n_anchors"]]

# Round all values in the mean table to 2 decimal places
mean_table_noHE.iloc[:, 1:] = mean_table_noHE.iloc[:, 1:].round(2)
```

```python
mean_table_noHE
```

```python
# Set the 'n_anchors' as the index for visualization
heatmap_data = mean_table_noHE.set_index("n_anchors")

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap="Blues", fmt=".2f", cbar=True)
plt.title("Heatmap of Mean training time Across 3 Training Runs", fontsize=14, pad=25)
plt.ylabel("n_anchors")
plt.show()
```

```python
# Find the lowest value and its corresponding cell
min_value_noHE = mean_table_noHE.iloc[:, 1:].min().min()
min_cell = mean_table_noHE.iloc[:, 1:].stack().idxmin()  # Get the index of the minimum value

# Extract the corresponding n_anchors and window_size
n_anchors, window_size = min_cell

# Display the result
min_value_noHE, (n_anchors, window_size)
```
